```jupyter notebook```

It will usually just open up a browser window for you, but if it doesn't copy and paste that URL into your browser.


You can navigate your file system from this screen, create new directories and create new notebooks by clicking on new -> Python 3


You can change the title of the notebook by clicking "Untitled"  Code is organized by cells. You can create a new cell by clicking +.  The cell default is to run python code.  You can run code by pushing shift+enter.  Output is displayed below the cell

Cells can also be configured for markdown by going to Cell> Cell Type > Markdown. This is a convenient way to write instructions

You can also plot things directly in the notebook by using the magic command %matplotlib inline

````python
import numpy as np
import seaborn as sns
%matplotlib inline
sns.set(style="ticks")

rs = np.random
x = rs.gamma(2, size=1000)
y = -.5 * x + rs.normal(size=1000)

sns.jointplot(x, y, kind="hex", color="#4CB391")
````
